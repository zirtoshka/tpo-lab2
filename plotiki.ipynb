```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt

def process_csv_files():
    csv_files = glob.glob("*.csv")

    for csv_file in csv_files:
        df = pd.read_csv(csv_file, header=None, names=["x", "y"])

        df = df.dropna()
        df = df[df["y"].abs() <= 25]

        plt.figure()
        plt.scatter(df["x"], df["y"], color='blue', marker='o', s = 3)
        plt.xlabel("X")
        plt.ylabel("Y")
        plt.title(f"  {csv_file}")
        plt.grid()

        png_filename = os.path.splitext(csv_file)[0] + ".png"
        plt.show()
        plt.close()

    print("!")

if __name__ == "__main__":
    process_csv_files()
```
